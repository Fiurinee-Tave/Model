## 최종데이터셋 구축하기
> "농촌진흥청 국립원예특작과학원 오늘의 꽃" 데이터 + "순천만국가정원 탄생화" 데이터
- 날짜의 경우, "월"만 이용하여 사용 ("일"까지 가져가는 게 큰 의미가 없다고 판단)
- 중복 데이터의 경우, 여러 개의 꽃말을 지니거나 할 수도 있고 다른 월(달)을 나타낼 수 있기에 제거없이 진행

```python
df1 = pd.read_csv("C:/Users/pc/Desktop/민지/동아리/프로젝트(13기)/오늘의 꽃_최종.csv")
df1.drop(["Unnamed: 0", "root__resultCode", "root__resultMsg", "root__result__fDay", "root__repcategory", "root__result__dataNo", "root__result__fSctNm", "root__result__fEngNm", "root__result__fGrow", "root__result__fType", "root__result__publishOrg"], axis=1, inplace=True) #일자, 학명, 영어이름, 기르기, 자생지, 출처 칼럼 제거
df1.drop(["root__result__fileName1", "root__result__fileName2", "root__result__fileName3", "root__result__imgUrl1", "root__result__imgUrl2"], axis=1, inplace=True) #이미지 1개만 남기기

df2 = pd.read_csv("C:/Users/pc/Desktop/민지/동아리/프로젝트(13기)/순천만국가정원_탄생화.csv")
df2.drop("일", axis=1, inplace=True) #일자 칼럼 제거
```

```python
#칼럼명 변경
new_columns = ["월", "꽃", "꽃말", "내용", "이용", "이미지"]
df1 = df1.set_axis(new_columns, axis=1)

#내용, 이용 합치기
df1['설명']=df1['내용']+df1['이용']
df1.drop(['내용','이용'], axis=1, inplace=True)

#df2의 꽃말 뜻 -> 설명
df2 = df2.rename(columns={"꽃말 뜻":"설명"})
```

```python
#데이터 합치기
df = pd.concat([df1, df2], axis=0)

#월 순으로 정렬
df = df.sort_values("월", ascending=True) 

df.shape
```

```python
df_copy = df.copy()
df_copy.drop("일", axis=1, inplace=True)

df_copy = df_copy.sort_values("월", ascending=True) #날짜(월)순으로 정렬
```
