## 최종데이터셋 구축하기
> "농촌진흥청 국립원예특작과학원 오늘의 꽃" 데이터 + "순천만국가정원 탄생화" 데이터 + “어니스트 플라워”의 꽃도감 데이터
- 나무 및 꽃이 없는 데이터 삭제
- 날짜의 경우, "월"만 이용하여 사용 ("일"까지 가져가는 게 큰 의미가 없다고 판단)
- 날짜를 통해 "계절" 칼럼 추출
- 중복 데이터의 경우, 여러 개의 꽃말을 지니거나 할 수도 있고 다른 월(달)을 나타낼 수 있기에 제거없이 진행

### 오늘의 꽃 데이터 전처리

```python
#나무데이터 제거된 오늘의 꽃 파일
df1 = pd.read_csv("C:/Users/pc/Desktop/민지/동아리/프로젝트(13기)/오늘의 꽃_최종(나무제거).csv")
df1.drop(["Unnamed: 0", "Column1", "root__resultCode", "root__resultMsg", "root__result__dataNo", "root__result__fDay", "root__repcategory", "root__result__fSctNm", "root__result__fEngNm", "root__result__fGrow", "root__result__fType", "root__result__publishOrg", "Unnamed: 22", "Unnamed: 23"], axis=1, inplace=True) #일자, 학명, 영어이름, 기르기, 자생지, 출처 칼럼 제거
df1.drop(["root__result__fileName1", "root__result__fileName2", "root__result__fileName3", "root__result__imgUrl1", "root__result__imgUrl2"], axis=1, inplace=True) #이미지 1개만 남기기
df1.head(5)
```

```python
#칼럼명 변경
new_columns = ["월", "꽃", "꽃말", "내용", "이용", "이미지"]
df1 = df1.set_axis(new_columns, axis=1)

#내용, 이용 합치기
df1['설명']=df1['내용']+df1['이용']
df1.drop(['내용','이용'], axis=1, inplace=True)
```

```python
print(df1.shape)
df1.head(5)
```

### 순천만 탄생화 데이터 전처리

```python
#나무데이터 제거된 순천만 탄생화 파일
df2_1 = pd.read_csv("C:/Users/pc/Desktop/민지/동아리/프로젝트(13기)/순천만_꽃_나무제거1.csv")
df2_1.drop("Unnamed: 0", axis=1, inplace=True)
df2_2 = pd.read_csv("C:/Users/pc/Desktop/민지/동아리/프로젝트(13기)/순천만_꽃_나무제거2.csv")
df2_2.drop("Unnamed: 0", axis=1, inplace=True)
```

```python
df2 = df2_1.merge(df2_2, on=["꽃","꽃말","월","설명","이미지"], how="outer")
print(df2.shape)
df2.head(5)
```

### 꽃도감 데이터 전처리

```python
df3 = pd.read_csv("C:/Users/pc/Desktop/민지/동아리/프로젝트(13기)/웹크롤링_꽃_최종2(나무제거).csv", encoding="cp949")
df3.drop("Unnamed: 0", axis=1, inplace=True)
df3.head(5)
```

```python
df3.shape
```

### 데이터 합치기

```python
df = df1.merge(df2, on=["꽃","꽃말","월","설명","이미지"], how="outer")
df = pd.concat([df, df3], ignore_index=True)
print(df.shape)
df.head(5)
```

```python
#계절 칼럼 채우기
#겨울(12,1,2) / 봄(3,4,5) / 여름(6,7,8) / 가을(9,10,11)

# 조건에 따라 계절 값 할당
def assign_season(month):
    if month in [12, 1, 2]:
        return '겨울'
    elif month in [3, 4, 5]:
        return '봄'
    elif month in [6, 7, 8]:
        return '여름'
    else:
        return '가을'

# NaN인 경우에만 계절 할당
df.loc[df['계절'].isna(), '계절'] = df.loc[df['계절'].isna(), '월'].apply(assign_season)
print(df.shape)
df.head(5)
```

```python
# 원하는 칼럼 순서로 재배치
new_column_order = ['꽃', '월', '계절', '꽃말', '설명', '이미지']
result_df = df.reindex(columns=new_column_order)
result_df = result_df.sort_values(["월","꽃"], ascending=True) #월, 꽃 순으로 재배치
result_df.head(5)
```
